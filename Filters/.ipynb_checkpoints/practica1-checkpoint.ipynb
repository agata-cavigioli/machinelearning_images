# ejercicio 1
Utlizando la imagen1.jpg, crea un vídeo en el que se vaya escalando y rotando la imagen (transformación de similaridad) en el punto del centro. El escalado inicial deberá ser igual a 0.1 y el ángulo inicial de 0º. Haz un total de 36 iteraciones aumentando la escala en 0.2 y el ángulo en 15º en cada iteración. Guardar el vídeo en el directorio de salida como video1.mp4. 

```python
import cv2
import numpy as np

image_path = 'imagen1.jpg'
img = cv2.imread(image_path)

output_video_path = 'salida/video1.mp4'
scale = 0.1
angle = 0
iterations = 36



height, width, channels = img.shape
center = (width // 2, height // 2)

fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_video_path, fourcc, 30, (width, height))

for _ in range(iterations):
    M = cv2.getRotationMatrix2D(center, angle, scale)
    rotated = cv2.warpAffine(img, M, (width, height))

    out.write(rotated)

    angle += 15
    scale += 0.2

out.release()
cv2.waitKey(0)
cv2.destroyAllWindows()


def showvideo(output_video_path):
    cap = cv2.VideoCapture(output_video_path)
    if not cap.isOpened():
        print("Error al abrir el archivo de video.")
        exit()

    while True:
        ret, frame = cap.read()
        if not ret:
            break
        cv2.imshow('Video', frame)
        if cv2.waitKey(30) & 0xFF == ord('q'):  # Presiona 'q' para salir
            break

    cap.release()
    cv2.destroyAllWindows()
    
showvideo(output_video_path) # Presiona 'q' para salir
```

# ejercicio 2
Intercambia el canal R con el canal B de la imagen2.png y pon a cero el canal G. Posteriormente, crea un video que vaya mezclando las dos imágenes por filas, de forma que aparezca primero la imagen con los canales intercambiados y vaya apareciendo poco a poco la imagen original. El número de iteraciones debe ser igual al alto de la imagen y
en cada una se debe ir desplazando una fila hacia abajo la imagen original sobre la transformada. Guarda el vídeo con el nombre vídeo2.mp4.

```python
image_swap = cv2.imread('imagen2.png')
image_original = cv2.imread('imagen2.png')

image_swap[:, :, 0], image_swap[:, :, 2] = image_swap[:, :, 2], image_swap[:, :, 0]
image_swap[:, :, 1] = 0

height, width, _ = image_original.shape

output_video_path = 'salida/video2.mp4'
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_video_path, fourcc, 30, (width, height))

iterations = height
mixed_image = image_swap

for i in range(iterations):
    mixed_image[i] = image_original[i]

    out.write(mixed_image)

out.release()
cv2.waitKey(0)
cv2.destroyAllWindows()

showvideo(output_video_path) # Presiona 'q' para salir
```

# ejercicio 3
Ecualiza el histograma de la imagen imagen3.png y luego elimina el ruido utilizando
un filtro mediano con un radio (ksize) igual a 8. Guarda el resultado como
imagen3.png.

```python
import matplotlib.pyplot as plt

image_path = 'imagen3.png'
img = cv2.imread(image_path)

output_img_path = 'salida/imagen3.png'

hist = cv2.calcHist([img], [0], None, [256], [0, 256])

plt.figure()
plt.axis("off")
plt.imshow(img, cmap='gray')

#not normalized
plt.figure()
plt.subplot(1, 2, 1)
plt.title("Histograma")
plt.xlabel("Bins")
plt.ylabel("# píxeles")
plt.plot(hist)
plt.xlim([0, 256])

#filtro mediano con un radio de 8 (usando 17 como valor impar)
filtered_image = cv2.medianBlur(img, 9)

plt.figure()
plt.axis("off")
plt.imshow(filtered_image, cmap='gray')


cv2.imwrite('salida/imagen_filtrada.png', filtered_image)
```

# ejercicio 4

Con la imagen4.jpg, crea un vídeo de 100 frames aplicando un filtro bilateral con
diámetro d=9 y valores de sigma de 11 y 11. Aplica el filtro en cada iteración al resultado
de la iteración anterior. Guarda el vídeo con el nombre video4.mp4

```python
image_path = 'imagen4.jpg'
img = cv2.imread(image_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

output_video_path = 'salida/video4.mp4'

plt.figure()
plt.axis("off")
plt.imshow(img)

height, width, channels = img.shape

fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_video_path, fourcc, 30, (width, height))

iterations = 100
sigma_color = 11
sigma_space = 11
new_img = img

for _ in range(iterations):
    new_img = cv2.bilateralFilter(img, d=9, sigmaColor=sigma_color, sigmaSpace=sigma_space)
    new_img = cv2.cvtColor(new_img, cv2.COLOR_BGR2RGB)
    out.write(new_img)

out.release()
    
showvideo(output_video_path) # Presiona 'q' para salir
```

# ejercicio 5

Calcula la norma del gradiente de la imagen5.png utilizando los filtros de las
derivadas centradas (-0.5, 0, 0.5). Umbraliza los valores de la norma de forma que los
valores mayores o iguales a 200 se igualen a 255, los menores de 100 se igualen a 0 y los
valores comprendidos entre 100 y 200 se igualen a 128. Finalmente, aplica un filtro
bilateral de radio 7 y con sigma -1. Guarda el resultado con el nombre imagen5.png.

```python
image = cv2.imread('imagen5.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.figure()
plt.axis("off")
plt.imshow(image)

grad_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
grad_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)

gradient_magnitude = cv2.magnitude(grad_x, grad_y)
gradient_magnitude = cv2.normalize(gradient_magnitude, None, 0, 255, cv2.NORM_MINMAX, cv2.CV_8U)
plt.figure()
plt.axis("off")
plt.imshow(gradient_magnitude)

gradient_magnitude[(gradient_magnitude >= 200)] = 255
gradient_magnitude[(gradient_magnitude < 100)] = 0
gradient_magnitude[(gradient_magnitude >= 100) & (gradient_magnitude < 200)] = 128

bilateral_filtered_image = cv2.bilateralFilter(gradient_magnitude, d=7, sigmaColor=-1, sigmaSpace=7)

cv2.imwrite('salida/imagen5.png', bilateral_filtered_image)
plt.figure()
plt.axis("off")
plt.imshow(bilateral_filtered_image)
```

# ejercicio 7
Mezcla las imágenes imagen7_1.jpg e imagen7_2.png de forma no lineal
u�lizando la siguiente expresión ... Crea un vídeo de 360 frames aumentando el ángulo en 1o en cada iteración y guárdalo
con el nombre de video7.mp4.

```python
import math

image1 = cv2.imread('imagen7_1.jpg')
plt.figure()
plt.axis("off")
plt.imshow(image1)
image2 = cv2.imread('imagen7_2.png')
plt.figure()
plt.axis("off")
plt.imshow(image2)

if image1 is None or image2 is None:
    print("Error al cargar una de las imágenes.")
    exit()

# Convert the PIL images to OpenCV format
image1 = cv2.cvtColor(np.array(image1), cv2.COLOR_RGB2BGR)
image2 = cv2.cvtColor(np.array(image2), cv2.COLOR_RGB2BGR)

output_video_path = 'salida/video7.mp4'
height, width, channels = image1.shape
center = (width // 2, height // 2)
total_frames = 360

fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_video_path, fourcc, 30, (width, height))

for angle in range(total_frames):
    alpha = angle * math.pi / 90.0 

    term1 = 0.5 * (1 + math.sin(alpha))
    term2 = 0.5 * (1 - math.sin(alpha))
    blended_image = cv2.addWeighted(image1, term1, image2, term2, 0)

    out.write(blended_image)

out.release()
cv2.destroyAllWindows()

showvideo(output_video_path) # Presiona 'q' para salir


```
