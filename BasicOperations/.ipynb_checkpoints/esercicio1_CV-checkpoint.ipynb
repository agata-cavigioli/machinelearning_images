```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
ancho = 200
alto = 300
gris_img = np.zeros((ancho,alto,1),dtype = np.uint8)
print(gris_img.shape)
cv2.imshow('color image',gris_img)
cv2.waitKey(1)
cv2.destroyAllWindows()
```

```python
plt.imshow(gris_img, cmap='gray')
plt.show()
```

```python
gris_img[:] = 10
plt.imshow(gris_img, cmap='gray')
plt.show()
```

```python
ancho = 800
alto = 800
gris_img = np.zeros((ancho,alto,1),dtype = np.uint8)
for n in range(0,9):
    for m in range(0,9):
        if n%2 == m%2:
            gris_img[0+(n*100):100+(n*100),0+(m*100):100+(m*100),0] = 255

plt.imshow(gris_img, cmap='gray')
plt.show()
```

```python
ancho = 800
alto = 800
img = np.zeros((ancho,alto,3),np.uint8) 
#img[0:100,0:100,2] = 255
#rgb
blanco = (255, 255, 255)
rojo = (255,0,0)

img = cv2.rectangle(img, (0, 0), (220, 300), blanco, -1)
img = cv2.rectangle(img, (250, 0), (800, 300), rojo, -1)

img = cv2.rectangle(img, (0, 330), (220, 500), blanco, -1)
img = cv2.rectangle(img, (250,330), (500, 500), blanco, -1)

plt.imshow(img)
plt.show()
```
