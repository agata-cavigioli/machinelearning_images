```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

                          
file = 'eii.png'
img = cv.imread(file)
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.figure()
plt.axis("off")
plt.imshow(img) 
plt.show()


#######################MEDIA
kernel = np.ones((11,1),np.float32)/11
fil = cv.filter2D(img, -1, kernel)
plt.figure()
plt.axis("off")
plt.imshow(fil) 
plt.show()

######################BLUR
avg = cv.blur(img,(5,5))
plt.figure()
plt.axis("off")
plt.imshow(avg) 
plt.show()

#############GAUSSIANBLUR
gaus = cv.GaussianBlur(img,(5,5),10)
plt.figure()
plt.axis("off")
plt.imshow(gaus) 
plt.show()
```

```python
file = 'figuras.png'
img = cv.imread(file)

##################FILTRADO LINEAL-DERIVADAS PARCIALES

kernel = np.cross(np.array([-0.5,0,0.5],float),np.array([-0.5,0,0.5],float).T)
print(np.array([-0.5,0,0.5],float))
kernel = np.array([-0.5,0,0.5]).T
#print(np.array([-0.5,0,0.5],float))
print(kernel)
fil = cv.filter2D(img, -1, kernel)
plt.figure()
plt.axis("off")
plt.imshow(fil) 
plt.show()

    #######################sobel
sob = cv.Sobel(img, -1, 1,0)
plt.figure()
plt.axis("off")
plt.imshow(sob) 
plt.show()


    #######################sobel
sob = cv.Laplacian(img, -1)
plt.figure()
plt.axis("off")
plt.imshow(sob) 
plt.show()
```
