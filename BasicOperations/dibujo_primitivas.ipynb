```python
import cv2
import numpy as np

def nothing(x):
    pass

#lineas, rectangulos, circulos
img = np.zeros((512, 512, 3), dtype=np.uint8)
img_copy = img.copy()

cv2.namedWindow('Drawing Shapes')

# Create trackbars

cv2.createTrackbar('B', 'Drawing Shapes', 255, 255, nothing)
cv2.createTrackbar('G', 'Drawing Shapes', 0, 255, nothing)
cv2.createTrackbar('R', 'Drawing Shapes', 0, 255, nothing)

# global variables
drawing = False  # True if mouse is pressed
mode = 'r'  # r para dibujar rectángulos, c para dibujar círculos, l para dibujar círculos
ix, iy = -1, -1
b = 0
g = 0
r = 255
cursor_color = (b, g, r)

def draw_shape(event, x, y, flags, param):
    global ix, iy, drawing, mode,img,img_copy,out,cursor_color,b,g,r

    if event == cv2.EVENT_LBUTTONDOWN:
        drawing = True
        ix, iy = x, y
        
    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing:
            img = img_copy.copy()
            if mode == 'r':
                cv2.rectangle(img, (ix, iy), (x, y), cursor_color, 2)
            elif mode == 'c':
                cv2.circle(img, (ix, iy), abs(x-ix + y-iy), cursor_color, 2)
            elif mode == 'l':
                cv2.line(img, (ix, iy), (x, y), cursor_color, 2)

    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False
        if mode == 'r':
            cv2.rectangle(img, (ix, iy), (x, y), cursor_color, -1)
        elif mode == 'c':
            cv2.circle(img, (ix, iy), abs(x-ix + y-iy) ,cursor_color, -1)
        elif mode == 'l':
            cv2.line(img,(ix, iy),(x, y),cursor_color,5)
        img_copy = img.copy()
        
    out.write(img)


#cv2.namedWindow('Drawing Shapes')
cv2.setMouseCallback('Drawing Shapes', draw_shape)
    

# Configurar la grabación del video
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('drawing_video.avi', fourcc, 20.0, (512, 512))

print('press r to draw rectangles, c to draw circles, l to draw lines')

while True:
    cv2.imshow('Drawing Shapes', img)
    k = cv2.waitKey(1) & 0xFF
    if k == ord('c'):
        mode = 'c'
    elif k == ord('l'):
        mode = 'l'
    elif k == ord('r'):
        mode = 'r'
    elif k == 27:  # esc to close the window
        break
    b = cv2.getTrackbarPos('B','Drawing Shapes')
    g = cv2.getTrackbarPos('G','Drawing Shapes')
    r = cv2.getTrackbarPos('R','Drawing Shapes')
    cursor_color = (b,g,r)


out.release()
cv2.destroyAllWindows()

```
